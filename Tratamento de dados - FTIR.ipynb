## Análise do espectro do Óxido de Grafeno - FTIR

```python
df_FTIR_GO = pd.read_fwf("FTIR_GO.txt")
```

```python
df_FTIR_GO['y'] = df_FTIR_GO['y'] - df_asu['y'] + 100
```

```python
df_FTIR_GO.head(137)
```

```python
# Remoção de outlier
for i in range(len(df_FTIR_GO['y'])):
    y = df_FTIR_GO['y'][i]
    x = df_FTIR_GO['x'][i]
    if y < 20:
        print(y)
        print(x)
        print(i)
        df_FTIR_GO = df_FTIR_GO.drop([i], axis=0)
```

```python
df_FTIR_GO.head(137)
```

```python
fig10= plt.figure()
ax10 = fig10.add_subplot(111)
df_FTIR_GO.plot('x','y',legend=True, label = 'Amostra do GO', ax=ax10, kind = 'line', linewidth = 1, color = 'blue')
plt.title('Caracterização do Óxido de Grafeno - FTIR')
plt.xlabel('Número de onda (1/cm)')
plt.ylabel('Transmitância (%)')
plt.savefig('Caracterização do Óxido de Grafeno - FTIR.png', dpi=100, format='png', facecolor='white')
plt.show()
```
